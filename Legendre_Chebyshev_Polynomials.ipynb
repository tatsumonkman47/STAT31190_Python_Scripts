```python
import matplotlib.pyplot as plt
from matplotlib import gridspec
import numpy as np
import math
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
```

```python
"""
Legendre, Chebyshevs, and Chebyshev derivative polynomials derived a couple of ways

"""
n = 10 # choose how high you want to go

x = np.arange(-1,1.01,0.01)

# the first few legendre polynomials written out after looking at wikipedia
lp = []
lp.append(x*0+1)
lp.append(x)
lp.append((1/2)*(3*x**2 - 1))
lp.append((1/2)*(5*x**3 - 3*x))
lp.append((1/8)*(35*x**4 - 30*x**2 + 3))
lp.append((1/8)*(63*x**5 - 70*x**3 + 15*x))
lp.append((1/16)*(231*x**6 - 315*x**4 + 105*x**2 -5))

# the first few legendre polynomials of the first kind via a generating function
lp_1gen = []
lp_1gen.append(x*0+1)
lp_1gen.append(x)
for i in range(2,n):
    lp_1gen.append(((2*i-1)*x*lp_1gen[i-1] - (i-1)*lp_1gen[i-2])/i)

# the first few chebyshev polynomials of the first kind via a generating function
cp1 = []
cp1.append(x*0+1)
cp1.append(x)
# the first derivatives of each chebyshev polynomial 
# taken from https://scicomp.stackexchange.com/questions/28309/derivatives-of-a-chebychev-polynomial
qp1 = []
qp1.append(x*0)
qp1.append(x*0+1)
for i in range(2,n):
    cp1.append(2*x*cp1[i-1] - cp1[i-2])
    qp1.append(2*cp1[i-1] + 2*x*qp1[i-1] - qp1[i-2])    
    
# the first few chebyshev polynomials of the second kind via a generating function
cp2 = []
cp2.append(x*0+1)
cp2.append(x)
for i in range(2,n):
    cp2.append(4*x*cp2[i-1] - cp2[i-2])

```

```python
fig = plt.figure(figsize=(22,11))
gs = gridspec.GridSpec(nrows=2,ncols=2)
ax00 = fig.add_subplot(gs[0,0])
ax01 = fig.add_subplot(gs[0,1])
ax10 = fig.add_subplot(gs[1,0])
ax11 = fig.add_subplot(gs[1,1])



ax00.set_title("Legendre Polynomials of the first kind")
for i in range(0,n):
    ax00.plot(x,lp_1gen[i],label=str(i))
ax00.legend()
ax00.grid()

ax01.set_title("Chebyshev Polynomials of the first kind")
for i in range(0,n):
    ax01.plot(x,cp1[i],label=str(i))
ax01.legend()
ax01.grid()

ax10.set_title("Derivatives of Chebyshev Polynomials of the first kind")
for i in range(0,n):
    ax10.plot(x,qp1[i],label=str(i))
ax10.set_ylim(-7,7)
ax10.legend()
ax10.grid()

ax11.set_title("Chebyshev Polynomials of the second kind")
for i in range(0,n):
    ax11.plot(x,cp2[i],label=str(i))
ax11.set_ylim(-5,5)
ax11.legend()
ax11.grid()
```
