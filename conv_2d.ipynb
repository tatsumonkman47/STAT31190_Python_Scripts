```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
```

```python
# a simple script for 2d convolutions for two nxn matrices

# initialize
n = 40 # size of matrices
x = np.random.rand(n,n) # data 
amat = np.random.rand(n,n) # operator
```

```python
# test the problem

y = np.zeros(x.shape)

# brute force for-loop implementation (loop over each row and column)
for i in range(0,n):
    for j in range(0,n):
        ytemp = 0
        for ii in range(0,n):
            for jj in range(0,n):
                idiff = i-ii
                jdiff = j-jj
                # simple way to wrap edges
                if (idiff < 0):
                    idiff = idiff + n
                if (jdiff < 0):
                    jdiff = jdiff + n
                ytemp = ytemp + amat[idiff,jdiff]*x[ii,jj]
        y[i,j] = ytemp
# end for loops
                
```

```python
y_fft = signal.fftconvolve(x,amat,mode="same")
```

```python
error_fft_vs_for = np.linalg.norm(y_fft-y)
```

```python
plt.pcolor(y_fft)
plt.colorbar()
plt.show()
plt.close()
plt.pcolor(y)
plt.colorbar()
plt.show()
plt.close()
```
